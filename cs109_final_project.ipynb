**Training, Validation, and Test Sets**

Kaggle provides 30 days of log data. The first 27 days are given as training, and the last 3 days are for testing (prediction). Competitors submit rerankings for test queries embedded in the test set, and Kaggle return a single summary score - an average NDCG score over all test queries. As actions that take place after a test query are not included in the provided test set, we would have had no real way of obtaining a finer grained understanding of how well our methods were performing.

This necessitated the creation of our own training, validation, and test sets from the training set that was provided to us. Essentially what this involved was choosing queries from the validation and test sets to be test queries following a sampling procedure.

More specifically, the sampling algorithm proceeded as follows:

<ol>
<li>Extract all log data corresponding to the first N users.</li>
<li>Separate the data into three sets: training (Days 0-19), validation (Days 20-22), and test (Days 23-25).</li>
<li>For both the validation and test sets, identify all queries that could potentially be valid test queries. A query qualifies if it meats the following two conditions:
<ul>
    <li> The user that made the query has activity in the training set.
    <li> Of the urls that are returned by the search engine, at least one is clicked on with a dwell time of 50 or more time units. 
</ul>
<li>For each user that ends up having potential test queries, choose one of those queries uniformly at random to become a test query.
</ol>

Another motivation for undertaking this task was to reduce our data to a more manageable size. We created 3 sets of (training, validation, test) files, corresponding to 1000, 10000, and 100000 users.

The code that accomplishes this task is in make_tvt.py.

###Evaluation Metric

We use the same metric as Kaggle- Normalized Discounted Cumulative Gain, or NDCG, to evaluate the effectiveness of our reranked search results.

URLs are assigned a *relevance* score based on if they were clicked and the dwell time of those clicks. Relevance scores fall into three buckets :

* 0 for URLs with no clicks and clicks with dwell time < 50 time units
* 1 for URLs with clicks with dwell time [50, 400) time units
* 2 for URLs with clicks with dwell time >= 400 time units and for clicks that correspond to the last action in the session

For any ranking of $p$ URLs, we can then calculate the DCG score as follows. $rel_i$ denotes the relevance of the URL at the *i*th position.

$$\mathrm{DCG_{p}} = rel_1 + \sum_{i=2}^{p} \frac{ {rel_{i}} }{ \log_{2}(i)}$$

We also calculate IDCG, the Ideal DCG score, which is the DCG score of the "optimal" ranking. The optimal ranking is simply a ranking of the URLs sorted by relevance score in descending order. 

NDCG is simply the DCG score divided by the IDCG score.

$$\mathrm{NDCG} = \frac{\mathrm{DCG}}{\mathrm{IDCG}}$$

Note that for this metric there isn't necessarily a single ideal ranking - potentially several rankings can achieve the perfect NDCG score of 1.

We take the average NDCG score over all test queries as our final summary metric.

The evaluation metric code is in eval.py. The two main functions are get_rel_map, which returns the relevance scores of all URLs that are to be ranked in the validation/test set by session, and compute_overall_score, which compute the summary average NDCG score given rel_map and a file with the proposed ranking of the URLs.

Examples of how the evaluation metric functions are used are in the next section.

###Latent Factor Model

As described in HW 4, latent factor models allow for identifying similar users and similar items as a function of latent factors and users' preferences for those latent factors. This kind of framework was very attractive to us as our problem can be thought of in a similar fashion. URLs, or websites, have certain characteristics (genre, topics language, amount of graphics, etc), and different users have different preferences for those characteristics.

The setup of our problem is almost identical to that presented in HW 4, except that instead of Yelp stars we have dwell time. As dwell time is an indicator of utility, we can use predicted dwell times to rerank a list of URLs.

**Training**

 The dwell time for a particular user $u$ and url $m$, $Y_{um}$, is modeled as a generative process:

$$Y_{um} = \mu + \theta_{u}[0] + \gamma_{m}[0] + \theta_{u}[1:]^{\top}\gamma_{m}[1:] + \epsilon_{um}$$

$$\epsilon_{um} \sim N(0, \sigma)$$

where 

* $\mu$ is the overall dwell time mean over all clicks.
* $\theta_u[0]$ is the user specific bias for user $u$
* $\gamma_m[0]$ is the item specific bias.
* $\theta_u[1:]$ are user $u$'s preferences for the latent factors.
* $\gamma_m[1:]$ are the latent factors associated with url $m$.
* $\sigma$ is the residual variance of dwell times after the mean, bias terms, and latent factors have been taken into account.

$\theta_u$ and $\gamma_m$ are L dimensional quantities, where L is the number of latent factors

Our goal in the *training* phase is to learn probability distributions over the unknown parameters $\mu$, $\sigma$, $\gamma_m$ (for all urls $m$), and $\theta_u$ (for all users $u$) given the dwell times we do observe. For notational compactness let $\Gamma$ be a $M$ by $L+1$ matrix with each row being a $\gamma_m$, and $\Theta$ be a $U$ by $L+1$ matrix with each row being a $\theta_u$. The posterior distribution over $\mu$, $\sigma$, $\Gamma$, and $\Theta$ given $Y$, the observed dwell times, is $P(\mu, \sigma, \Gamma, \Theta \mid Y)$.

Continuing along the lines of HW 4, we used Gibbs sampling to explore the posterior probability distribution. Gibbs sampling works by iteratively sampling each unknown, holding the other unknowns constant. We omit the form of the draws from the distributions of $\mu$, $\sigma$, $\Gamma$, and $\Theta$ as they are identical to HW 4. 

**Prediction**

Every iteration of the Gibbs sampler results in a sample from the distributions of $\mu$, $\sigma$, $\Gamma$, and $\Theta$. We can think of a series of draws accumulated over enough iterations as a good approximation of the actual posterior distribution, and we take the posterior mean of each parameter as our final estimate.

In the *prediction* phase, we use our estimates of each parameter to predict $Y_{um}$, the dwell time of user $u$ on URL $m$ by simply plugging into the expression

$$Y_{um} = \mu + \theta_{u}[0] + \gamma_{m}[0] + \theta_{u}[1:]^{\top}\gamma_{m}[1:]$$

**Practical Issues**

Before the presentation of results, a few words concerning practical matters with the data are in order. In training up our model, we only looked at clicks on URLs in which dwell time was well defined, and ignore URLs that were never clicked on.

The latent factor model we describe can only predict $Y_{um}$ for user $u$ and URL $m$ if we've learned $\theta_u$ and $\gamma_m$ the training phase. We cannot make a prediction for a never before seen new user or URL. Note however, that the model is perfectly capable of predicting $Y_{um}$ for a user $u$ has never before seen URL $m$ if as long as we have $\theta_u$ and $\gamma_m$.

This turned out to be a significant issue. For about 90% of the URLs that appear in the validation set, we are unable to predict a dwell time due to not having $\gamma_m$ for that URL. Thus we decided on the following reranking procedure, ranking the URLs for which we do have information about first, and then breaking ties for the "unknowns" by looking at the default search engine ranking. Here's an example.

Let's say we are presented with 5 URLs [B, C, A, D, F], and we are able to predict that C has a dwell time of 100 and F has a dwell time of 75. We would then return the reranking of [C, F, B, A, D], as C had the highest score, followed by F, and B, A, and D are returned in the default order.

First we train the latent factor model. build_df is a helper function to build the dataframe of dwell time data. The code for running the Gibbs sampler is virtually the same as in HW 4, and is included in latent_factor.py.

```python
# builds a dataframe with each row being data (session id, url id, and dwell time) for a click
def build_df(filename):
    f = open(filename)
    fr = FileReader(f)
    
    session_to_user_map = {}
    
    dwell_time_map_user = defaultdict(list)
    
    click_data = []
    
    while True:
        line = fr.get_next_line()
        if not line:
            break
        elems = line.split()
        if elems[1] == 'M':
            metadata = parse_metadata(elems)
            session_id = metadata['session_id']
            user_id = metadata['user_id']
            session_to_user_map[session_id] = user_id
        elif elems[2] == 'C':
            click = parse_click(elems)
            session_id = click['session_id']
            user_id = session_to_user_map[session_id]
            url_id = click['url_id']
            time_passed = click['time_passed']
            
            next_line = fr.peek_line()
            # if this is the last click of the session
            next_line = fr.peek_line()
            # if this is the last click of the file
            if not next_line:
                break
            next_elems = next_line.split()
            # if this is the last click of the session
            if next_elems[1] == 'M':
                pass
            else:
                next_time_passed = int(next_elems[1])
                dwell_time = next_time_passed - time_passed
                if dwell_time > 0:
                    data = {'session_id': session_id, 'user_id': user_id, 'url_id': url_id, 'dwell_time': dwell_time}
                    click_data.append(data)
    
    f.close()
    df = pd.DataFrame(click_data)
    return df   
```

```python
def train_latent_factor(train_file, L=5, maxit=1000):
    df = build_df(train_file)
    
    users = np.unique(df['user_id'])
    urls = np.unique(df['url_id'])
       
    # create mappings between user id's and indexes
    user_idxs_map = {user: i for (i, user) in enumerate(users)}
    url_idxs_map = {url: i for (i, url) in enumerate(urls)}
    
    Lambda_theta_diag = 0.1
    Lambda_gamma_diag = 0.1
    
    gibbs_out = factor_gibbs(df, L, maxit, Lambda_theta_diag, Lambda_gamma_diag)
    
    return gibbs_out, user_idxs_map, url_idxs_map
```

Now for reranking. First, a few helper functions.

```python
# predict the dwell time given for a particular user and url
def predict_dwell_time(gibbs_out, user_idx, url_idx):
    mu, theta, gamma = get_parameters(gibbs_out, user_idx, url_idx)
    dwell_time = compute_dwell_time(mu, theta, gamma)
    return dwell_time

# compute dwell time prediction given parameters
def compute_dwell_time(mu, theta, gamma):
    dwell_time = mu + np.sum(theta[1:]*gamma[1:]) + theta[0] + gamma[0]
    return dwell_time

# obtain the parameter values for a particular user and url
def get_parameters(gibbs_out, idx_user, idx_url):
    burnin = 200
    mu = np.mean(gibbs_out['mu'][burnin:])
    theta_mat = np.mean(gibbs_out['theta'][:,:,burnin:], axis=2)
    gamma_mat = np.mean(gibbs_out['gamma'][:,:,burnin:], axis=2)
    
    theta = theta_mat[idx_user,:]
    gamma = gamma_mat[idx_url,:]
    
    return mu, theta, gamma

# rerank according to scores, keeping original ranking if scores are unavailable
def rerank(original_ranking, scores):
    score_ranking = sorted(range(len(scores)), key=lambda x:scores[x], reverse=True)
    reranked_ranking = [original_ranking[i] for i in score_ranking]
    return reranked_ranking

def write_ranking(f, session_id, ranked_urls):
    for url in ranked_urls:
        f.write(str(session_id) + ',' + str(url) + '\n')
```

Finally, the code for executing and creating a file containing a proposed reranking.

```python
def rerank_latent_factor(validate_file, user_idxs_map, url_idxs_map, gibbs_out):
    metadata, _, test_queries, _, _= get_sessions(validate_file)
    
    # create mapping from session_id to user_id
    session_map = {}
    for m in metadata:
        session_map[m['session_id']] = m['user_id']
    
    # create dataframes
    metadata_df = pd.DataFrame(metadata)
    test_queries_df = pd.DataFrame(test_queries)
    
    ranking_file = validate_file + '-latent-factor-ranking'
    f = open(ranking_file, 'w')
    f.write('SessionID,URLID\n')
    for test_query in test_queries:
        session_id = test_query['session_id']
        user = session_map[session_id]
        urls = zip(*test_query['list_of_urls_and_domains'])[0]
                
        if user not in user_idxs_map:
            write_ranking(f, session_id, urls)
            continue
        user_idx = user_idxs_map[user]
        
        url_idxs = [url_idxs_map[url] if url in url_idxs_map else None for url in urls]
        
        dwell_time_predictions = []
        for url_idx in url_idxs:
            #print url_idx
            if not url_idx:
                dwell_time_predictions.append(None)
                continue
            dwell_time_prediction = predict_dwell_time(gibbs_out, user_idx, url_idx)
            dwell_time_predictions.append(dwell_time_prediction)
        reranked_rankings = rerank(urls, dwell_time_predictions)
        write_ranking(f, session_id, reranked_rankings)
    f.close()
    
    return ranking_file
```

### Conclusions and Future Work

It became clear to us throughout the process of working on this project that the default ranking of URLs returned by the search engine is already very good. While the methods we use appear to improve on the default rankings ever so slightly, we can't claim that they are close to solving the search personalization problem.

The reranking models we use are objectively speaking fairly simple. One reason for that is the limited amount of data we have and the fact that it is all anonymized. One can imagine hundreds if not thousands of factors that would influence how long a user dwells on a particular website - ranging from the actual content of the website itself to time of day to how ADD a particular user is. In reality these could all be modeled somehow, but for this competition we are only really provided a tiny subset of those factors.

That being said, we didn't take full advantage of the limited data we were given. For example, none of our models take the temporal nature of the data into account. We also don't have a model for how to include short term context, i.e. actions (queries/clicks) that happen before the test query in the same session. These are potential areas for future work.

Finally, we only worked on a tiny subset of the entire training set provided by Kaggle (~0.2%). To scale up the models we use, we most certainly would have to use Map Reduce or other distributed computing techniques.

Overall, we found this project very interesting and a great learning experience in terms of dealing with big-ish data and carrying out a data science processing pipeline from start to finish.
